```python
#Overall Sales Trend (Business Health)
# Ensure date format
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
daily_sales = df.groupby('transaction_date')['final_amount'].sum()
plt.figure()
plt.plot(daily_sales)
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Top Product Categories by Revenue
category_sales = df.groupby('product_category')['final_amount'].sum().sort_values(ascending=False)
plt.figure()
category_sales.plot(kind='bar')
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#Quantity Sold vs Current Stock
plt.figure()
plt.scatter(df['quantity_sold'], df['current_stock'])
plt.title("Quantity Sold vs Current Stock")
plt.xlabel("Quantity Sold")
plt.ylabel("Current Stock")
plt.show()
```

```python
#Expiry Risk Analysis
# Product Expiry Risk Overview

df['expiry_date'] = pd.to_datetime(df['expiry_date'])
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

df['days_to_expiry'] = (df['expiry_date'] - df['transaction_date']).dt.days

df['expiry_status'] = pd.cut(
    df['days_to_expiry'],
    bins=[-1, 7, 30, df['days_to_expiry'].max()],
    labels=['Expiring in 7 Days', 'Expiring in 30 Days', 'Safe Stock']
)

expiry_summary = df['expiry_status'].value_counts()

plt.figure()
expiry_summary.plot(kind='pie')
plt.title("Product Expiry Risk Overview")
plt.xlabel("Expiry Category")
plt.ylabel("Number of Products")
plt.show()
```

```python
#Festival Impact on Sales
festival_sales = df.groupby('festival_flag')['final_amount'].sum()

plt.figure()
festival_sales.plot(kind='bar')
plt.title("Sales: Festival vs Non-Festival")
plt.xlabel("Festival Flag (0 = No, 1 = Yes)")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Discount Effectiveness 
plt.figure()
plt.scatter(df['discount_amount'], df['quantity_sold'])
plt.title("Discount vs Quantity Sold")
plt.xlabel("Discount Amount")
plt.ylabel("Quantity Sold")
plt.show()
```
